```python
from tensorflow.keras import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD
# from tensorflow.keras.optimizers import Adam
from keras.models import Sequential
import numpy as np
import pandas as pd
```

```python
# defining input and output
x1 = np.array([0, 0, 1, 1], dtype=float)
x2 = np.array([0, 1, 0, 1], dtype=float)
y = np.array([0, 1, 1, 0], dtype=float)
```

```python
model = Sequential(
    [
        Dense(8, activation="relu", name="layer1"),
        Dense(1, activation="sigmoid", name="layer2"),
        # layers.Dense(4, name="layer3"),
    ]
)
```

```python
# model.compile(
#    loss = 'binary_crossentropy', optimizer = SGD(learning_rate=0.01), metrics = ['accuracy']
# )

model.compile(
   loss = 'binary_crossentropy', optimizer = SGD(learning_rate=0.1,), metrics = ['accuracy']
)
```

```python
X = np.column_stack((x1, x2))
model.fit(X, y, epochs=100, verbose=0)
```

```python
model.evaluate(X,y)
```

```python
pred = model.predict(X)
print(pred)
```
